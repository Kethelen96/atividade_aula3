```
import statistics
empresa1 = [1000,6000,1200,8000,1400]
empresa2 = [5000,4000,3000,2000,7000]
empresa3 = [1200,1300,8000,3000,15000]
empresa4 = [1400,1750,2000,4500,5900,7000]
empresa5 = [1400,1750,2000,4500,5900,7000]

def moda():
    moda_empresa1 = statistics.mode(empresa1)
    print ("Moda empresa 1:", moda_empresa1)
    moda_empresa2 = statistics.mode(empresa2)
    print ("Moda empresa 2:", moda_empresa2)
    moda_empresa3 = statistics.mode(empresa3)
    print ("Moda empresa 3:", moda_empresa3)
    moda_empresa4 = statistics.mode(empresa4)
    print ("Moda empresa 4:", moda_empresa4)
    moda_empresa5 = statistics.mode(empresa5)
    print ("Moda empresa 5:", moda_empresa5)

moda()

def mediana():
    mediana_empresa1 = statistics.median(empresa1)
    print ("Mediana empresa 1:", mediana_empresa1)
    mediana_empresa2 = statistics.median(empresa2)
    print ("Mediana empresa 2:", mediana_empresa2)
    mediana_empresa3 = statistics.median(empresa3)
    print ("Mediana empresa 3:", mediana_empresa3)
    mediana_empresa4 = statistics.median(empresa4)
    print ("Mediana empresa 4:", mediana_empresa4)
    mediana_empresa5 = statistics.median(empresa5)
    print ("Mediana empresa 5:", mediana_empresa5)

mediana()

def media():
    media_empresa1 = statistics.mean(empresa1)
    print ("Media empresa 1:", media_empresa1)
    media_empresa2 = statistics.mean(empresa2)
    print ("Media empresa 2:", media_empresa2)
    media_empresa3 = statistics.mean(empresa3)
    print ("Media empresa 3:", media_empresa3)
    media_empresa4 = statistics.mean(empresa4)
    print ("Media empresa 4:", media_empresa4)
    media_empresa5 = statistics.mean(empresa5)
    print ("Media empresa 5:", media_empresa5)

media()

def desvio_padrao():
    desvio_empresa1 = statistics.stdev(empresa1)
    print ("Desvio padrão empresa 1:", desvio_empresa1)
    desvio_empresa2 = statistics.stdev(empresa2)
    print ("Desvio padrão empresa 2:", desvio_empresa2)
    desvio_empresa3 = statistics.stdev(empresa3)
    print ("Desvio padrão empresa 3:", desvio_empresa3)
    desvio_empresa4 = statistics.stdev(empresa4)
    print ("Desvio padrão empresa 4:", desvio_empresa4)
    desvio_empresa5 = statistics.stdev(empresa5)
    print ("Desvio padrão empresa 5:", desvio_empresa5)

desvio_padrao()

# Mostrar o resultado da função para a empresa 1
moda_result1 = statistics.mode(empresa1)
mediana_result1 = statistics.median(empresa1)
media_result1 = statistics.mean(empresa1)
desvio_result1 = statistics.stdev(empresa1)
# Usar os resultados para decidir ou não aceitar a empresa 1
if moda_result1 >= 4000:
    print("Aceitar proposta de emprego empresa 1")
elif mediana_result1 >= 4000:
    print("Aceitar proposta de emprego empresa 1")
elif media_result1 >= 4000:
    print("Aceitar proposta de emprego empresa 1")
elif desvio_result1 >= 4000:
    print("Aceitar proposta de emprego empresa 1")
else:
    print("Recusar proposta de emprego empresa 1")

# Mostrar o resultado da função para a empresa 2
moda_result2 = statistics.mode(empresa2)
mediana_result2 = statistics.median(empresa2)
media_result2 = statistics.mean(empresa2)
desvio_result2 = statistics.stdev(empresa2)
# Usar os resultados para decidir ou não aceitar a empresa 2
if moda_result2 >= 4000:
    print("Aceitar proposta de emprego empresa 2")
elif mediana_result2 >= 4000:
    print("Aceitar proposta de emprego empresa 2")
elif media_result2 >= 4000:
    print("Aceitar proposta de emprego empresa 2")
elif desvio_result2 >= 4000:
    print("Aceitar proposta de emprego empresa 2")
else:
    print("Recusar proposta de emprego empresa 2")

# Mostrar o resultado da função para a empresa 3
moda_result3 = statistics.mode(empresa3)
mediana_result3 = statistics.median(empresa3)
media_result3 = statistics.mean(empresa3)
desvio_result3 = statistics.stdev(empresa3)
# Usar os resultados para decidir ou não aceitar a empresa 3
if moda_result3 >= 4000:
    print("Aceitar proposta de emprego empresa 3")
elif mediana_result3 >= 4000:
    print("Aceitar proposta de emprego empresa 3")
elif media_result3 >= 4000:
    print("Aceitar proposta de emprego empresa 3")
elif desvio_result3 >= 4000:
    print("Aceitar proposta de emprego empresa 3")
else:
    print("Recusar proposta de emprego empresa 3")

# Mostrar o resultado da função para a empresa 4
moda_result4 = statistics.mode(empresa4)
mediana_result4 = statistics.median(empresa4)
media_result4 = statistics.mean(empresa4)
desvio_result4 = statistics.stdev(empresa4)
# Usar os resultados para decidir ou não aceitar a empresa 4
if moda_result4 >= 4000:
    print("Aceitar proposta de emprego empresa 4")
elif mediana_result4 >= 4000:
    print("Aceitar proposta de emprego empresa 4")
elif media_result4 >= 4000:
    print("Aceitar proposta de emprego empresa 4")
elif desvio_result4 >= 4000:
    print("Aceitar proposta de emprego empresa 4")
else:
    print("Recusar proposta de emprego empresa 4")

# Mostrar o resultado da função para a empresa 5
moda_result5 = statistics.mode(empresa5)
mediana_result5 = statistics.median(empresa5)
media_result5 = statistics.mean(empresa5)
desvio_result5 = statistics.stdev(empresa5)
# Usar os resultados para decidir ou não aceitar a empresa 5
if moda_result5 >= 4000:
    print("Aceitar proposta de emprego empresa 5")
elif mediana_result5 >= 4000:
    print("Aceitar proposta de emprego empresa 5")
elif media_result5 >= 4000:
    print("Aceitar proposta de emprego empresa 5")
elif desvio_result5 >= 4000:
    print("Aceitar proposta de emprego empresa 5")
else:
    print("Recusar proposta de emprego empresa 5")







```
